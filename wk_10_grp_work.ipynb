Write a function that calculates the dot product manually AND a function that
calculates the cross product manually. Both should accept two vectors and perform
that calculation. Be sure to use try except and error handling for basic issues that
might occur.

```python
def dot_prod(a, b):
    total = 0
    try:
        for i, n in zip(a, b):
            dot = i * n
            total = total + dot
    except:     
        print("There must e an error")
    return(total)

x = ([2, 6])
y = ([3, 7])

dot_prod(x, y)
```

```python
def cross_prod(a, b):
    c1 = 0
    c2 = 0
    c3 = 0
    try:
        c1 = a[1]*b[2] - a[2]*b[1]
        c2 = a[2]*b[0] - a[0]*b[2]
        c3 = a[0]*b[1] - a[1]*b[0]
    except:
        print("There appears to be a problem.")
    
    return np.array([c1, c2, c3])

x = ([1, 2, 3])
y = ([1, 1, 1])

cross_prod(x, y)
```

### Machine learning cannot work directly with categorical data. One hot encoding is a way to encode categorical data that does not follow an ordinal value, so that it can be used in algorithms.
